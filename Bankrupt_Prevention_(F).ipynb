Values close to 0 indicate a weak or no linear relationship.

```python
# Import the necessary libraries
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
import numpy as np
import matplotlib.pyplot as plt

# Assuming the data preprocessing code is here

# Model training part
clf = IsolationForest(random_state=15, contamination=0.015)
clf.fit(bankrupt_new)
y_pred_outliers = clf.predict(bankrupt_new)


x = bankrupt_new.iloc[:, :-2]
y = bankrupt_new['class_yn']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)

# Model evaluation part
classifiers = ['Naive Bayes', 'Logistic Regression', 'Random Forest', 'Gradient Boosting', 'Decision Tree']
accuracies = []
roc_auc_scores = []

for classifier_name in classifiers:
    if classifier_name == 'Naive Bayes':
        clf = MultinomialNB()
    elif classifier_name == 'Logistic Regression':
        clf = LogisticRegression()
    elif classifier_name == 'Random Forest':
        clf = RandomForestClassifier(random_state=0)
    elif classifier_name == 'Gradient Boosting':
        clf = GradientBoostingClassifier(random_state=0)
    elif classifier_name == 'Decision Tree':
        clf = DecisionTreeClassifier()

    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    accuracy = round(accuracy_score(y_test, y_pred) * 100, 2)
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    roc_auc = auc(fpr, tpr)

    print(f"{classifier_name} Accuracy: {accuracy}")
    print(classification_report(y_test, y_pred))
    print(f"Confusion Matrix for {classifier_name}:\n{confusion_matrix(y_test, y_pred)}")
    print(f"ROC AUC for {classifier_name}: {roc_auc}\n")

    accuracies.append(accuracy)
    roc_auc_scores.append(roc_auc)

# Create a bar plot for the accuracy of each classifier
y_pos = np.arange(len(classifiers))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(y_pos, accuracies, align='center', alpha=0.5)
plt.xticks(y_pos, classifiers)
plt.xlabel('Classifiers')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy of Different Classifiers on the Data Set')

for i in range(len(accuracies)):
    plt.text(x=i - 0.1, y=accuracies[i] + 0.5, s=str(accuracies[i]))

# Create a bar plot for the ROC AUC of each classifier
plt.subplot(1, 2, 2)
plt.bar(y_pos, roc_auc_scores, align='center', alpha=0.5)
plt.xticks(y_pos, classifiers)
plt.xlabel('Classifiers')
plt.ylabel('ROC AUC')
plt.title('ROC AUC of Different Classifiers on the Data Set')

for i in range(len(roc_auc_scores)):
    plt.text(x=i - 0.1, y=roc_auc_scores[i] + 0.005, s=f'{roc_auc_scores[i]:.3f}')

plt.tight_layout()

plt.show()
```
